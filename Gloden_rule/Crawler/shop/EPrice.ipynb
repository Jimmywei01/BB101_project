```python
from time import strftime
import requests
from bs4 import BeautifulSoup
import csv
#from selenium import webdriver
#from selenium.webdriver.common.keys import Keys
```

```python
# 單一標籤  
# title = soup.select_one("div.container.content > section.col-md-9 > div.search-rst.clearfix > h4.pd-name").text.strip()   
# price = soup.select_one("div.container.content > section.col-md-9 > div.search-rst.clearfix > span.pd-price > span.num").text
url = "https://ezprice.com.tw/s/生活百貨/price/?q=生活百貨&p="
now = strftime('%Y-%m-%d %H:%M:%S')
datall= []
for i in range(1,21):
    host = url+str(i)#換頁
    res = requests.get(host)
    soup = BeautifulSoup(res.text,"lxml")
    print(('Page'+str(i)+' ')*50) #顯示出現在爬到那一頁 讓我知道進度如何而已   


    for inners in soup.select('div.search-rst.clearfix'):
        data = '九元生活館',now,inners.select_one('h4.pd-name').text.strip(),inners.select_one('span.num').text+'元',' ',' ' #商品名 + 時間 + 價錢
        datall.append(data)
    print(datall)

headers = ['商家名稱(string)','資料時間(datetime)','商品名稱(string)','商品價格(var(9,2))','商品網址(string)','運費(var(6,2))']
 # 欄位設定
f = open('EPrice.csv','a') # 可用於str形式
w = csv.writer(f)
w.writerow(headers)
w.writerows(datall)
f.close()

# with open( 'hot3.csv', 'wb') as f: # 只能用於 list形式資料
#     writer = csv.writer(f)
#     writer.writerows(data)       
#     f.close()
```
