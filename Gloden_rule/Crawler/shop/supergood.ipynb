```python
from time import strftime
import requests
from bs4 import BeautifulSoup
import csv
#from selenium import webdriver
#from selenium.webdriver.common.keys import Keys
```

```python
now = strftime('%Y-%m-%d %H:%M:%S') # 現在時間
inners = []
for i in range(1,8):
    url ='http://www.ezp.com.tw/category.php?id=59&price_min=0&price_max=0&page='+str(i)+'&sort=shop_price&order=ASC#goods-list' #換頁
    res = requests.get(url)
    soup = BeautifulSoup(res.text,"lxml")
    print(('Page'+str(i)+' ')*50)  #顯示出現在爬到那一頁 讓我知道進度如何而已 
    for inner in soup.select('.info'):
        p = ' ',now,inner.select_one('.name').text,inner.select_one('em.price').text,' ',' '
        inners.append(p)
print(inners)


headers = ['商家名稱(string)','資料時間(datetime)','商品名稱(string)','商品價格(var(9,2))','商品網址(string)','運費(var(6,2))'] # 欄位設定
f = open('supergood.csv','a') # 可用於str形式
w = csv.writer(f)
w.writerow(headers)
w.writerows(inners)
f.close()

           
```
